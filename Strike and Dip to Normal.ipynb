```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

def fnormal(strike, dip):
    deg_to_rad = lambda x: x * np.pi / 180
    r_strike = deg_to_rad(strike)
    r_dip = deg_to_rad(dip)
    n = np.array([
        np.sin(r_dip) * np.cos(r_strike),
        -np.sin(r_dip) * np.sin(r_strike),
        -np.cos(r_dip)
    ])
    return n
```

```python
fnormal(0, 90)
```

```python
fnormal(30, 45)
```

```python
fnormal(280, 60)
```

```python
normal = fnormal(280, 60)
x = np.linspace(-10, 10)
y = np.linspace(-10, 10)

xx, yy = np.meshgrid(x, y)

z = -(normal[0] * xx + normal[1] * yy) / normal[2]

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(xx, yy, z, cmap ='viridis', edgecolor ='green')

```
